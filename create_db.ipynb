```python
import sqlite3 
import csv 
import pandas as pd
```

```python
def create_db(db_name,file_name,table_name):
    con = sqlite3.connect(db_name)
    cursor = con.cursor()

    Employee=pd.read_csv(file_name)
    Employee.to_sql(table_name,con,if_exists='replace',index=False)

    results=cursor.execute(f"select * from {table_name}").fetchall()
    con.commit()
    con.close()
    
    
    return(results)
```

```python
Game=create_db("Games1.db","dataset.csv","Game")
```
